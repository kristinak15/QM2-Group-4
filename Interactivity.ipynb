```
import pandas as pd
import matplotlib.pyplot as plt 
```

```
data = pd.read_csv('/content/IndexBase100 (1).csv')
```

```
data.plot(x='Years',y=['GM Corn Production Index Base 100','Public interactivity Index Base 100'])
plt.xlabel('Years')
plt.ylabel('Index 100')
plt.title('Effect of GM Corn Production on Public Interactivity on the topic')
fig = plt.gcf()
fig.set_size_inches(15, 8.5)
fig.savefig('interactivity-cornprod.png', dpi=100)
```
