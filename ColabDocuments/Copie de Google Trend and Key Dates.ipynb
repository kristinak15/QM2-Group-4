```
import pandas
import pylab
import numpy as np
import matplotlib.pyplot as plt
data = pandas.read_csv('./GoogleTrend.csv')
data.head()
```

```
#We plot the data
data.plot(x='Period', y='Interaction with the topic genetically modified crops in the United-States.')
#We then add the specific years which fueled the GM debate.
#The first event is the discovery of GM Toxins in human blood
plt.axvline(x=53, color="b")
#The second event is the first trial against Mosanto, a GM producer
plt.axvline(x=78, color="b")
#The third event is the first approval of FDA of a GM Salmon. 
plt.axvline(x=107, color="b")
#The fourth event is Congress passing a law to label GM food.
plt.axvline(x=115, color="b")
#The fifth event is the authorisation of GM apples for sales
plt.axvline(x=131, color="b")
#The sixth event is the consultation of the FDA to make food from genome-edited plants
plt.axvline(x=146, color="b")

#This entry is the online discussion of Chipotle decisions and lawsuits to get rid of GMO in their product.
plt.axvline(x=101, color="b")
#We adjust the graph size
plt.style.use('ggplot')
pylab.rcParams['figure.figsize'] = (30,20)


```
