```
import pandas
import pylab
import numpy as np
import matplotlib.pyplot as plt
data = pandas.read_csv('/content/cloned-repo/FINAL DATA SETS/GoogleTrend.csv')
data.head()
```

```
#We plot the data
data.plot(x='Period')
#We then add the specific years which fueled the GM debate.
#The first event is the discovery of GM Toxins in human blood
plt.axvline(x=55, color="r", label="Discovery of GM toxins in human blood", linewidth=4)
#The second event is the first trial against Mosanto, a GM producer
plt.axvline(x=78, color="y", label="First trial against Mosanta, GM producer", linewidth=4)
#The third event is the first approval of FDA of a GM Salmon. 
plt.axvline(x=107, color="c", label="Approval by the FDA of a GM salmon", linewidth=4)
#The fourth event is Congress passing a law to label GM food.
plt.axvline(x=115, color="b", label="Law to label all GM food", linewidth=4)
#The fifth event is the authorisation of GM apples for sales
plt.axvline(x=131, color="m", label="GM Apples are allowed for sales", linewidth=4)
#The sixth event is the consultation of the FDA to make food from genome-edited plants
plt.axvline(x=146, color="k", label ="Food from genome-edited plants discussion by the FDA", linewidth=4)

#This entry is the online discussion of Chipotle decisions and lawsuits to get rid of GMO in their product.
plt.axvline(x=101, color="g", label="Chipotle lawsuit after their decision not to use GM food in their products", linewidth=4)
#We adjust the graph size
l = plt.legend(loc ='upper left')
plt.style.use('ggplot')
pylab.rcParams['figure.figsize'] = (20,10)


```
