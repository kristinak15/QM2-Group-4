```
#This notebook will be used to run different analysis on the states level of interactivity with the topic of genetically modified crops.
#All data was formatted to the unit Index Base 100 - where a 100 represents the maximum value in the dataset.
#The data for the GM production is collected from the allTablesGECrops on Github and the data for public interactivity is from Google Trend

#Note that the datasets were merged on Numbers, as the formatting of Index Base 100, and the number of columns made it very difficult to do it on Python.

#We first upload the CSV file (this will be automatically done through Github later)
from google.colab import files
uploaded = files.upload()
```

```
#We import all possible Python libraries we might need to use.
import pandas
from pylab import rcParams
import pylab

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import scipy
from scipy.stats.stats import pearsonr
```

```
#We load the data
data = pandas.read_csv('./StatesIndex.csv')
data.head()
```

```
#We now compare the different level of production per states
states = ['Illinois genetically modified corn production','Indiana genetically modified corn production','Iowa genetically modified corn production','Kansas genetically modified corn production','Michigan genetically modified corn production','Minnesota genetically modified corn production','Missouri genetically modified corn production','Nebraska genetically modified corn production','Ohio genetically modified corn production','North Dakota genetically modified corn production','South Dakota genetically modified corn production','Texas genetically modified corn production','Wisconsin genetically modified corn production']
data.plot(x='Years',y= states)
plt.style.use('ggplot')
pylab.rcParams['figure.figsize'] = (20,10)
```

```
#We now plot on a graph the public interactivity online with the topic of GMO, for each states, throughout the period of 2007 and 2019
#Note that because of irregularities and absence of Google Trend data for Kansas, North Dakota, and South Dakota, those states will not be included.
interactivity = ['Public interactivity for Illinois','Public interactivity for Indiana','Public interactivity for Michigan','Public interactivity for Minnesota','Public interactivity for Iowa ','Public interactivity for Missouri','Public interactivity for Nebraska','Public interactivity for Ohio','Public interactivity for Texas','Public interactivity for Wisconsin']
data.plot(x='Years',y= interactivity)
plt.style.use('ggplot')
pylab.rcParams['figure.figsize'] = (20,10)
```

```
#We now calculate the Pearson correlation for all states, between their production of GM and the public engagement online with the topic
illinois1 = data["Illinois GM Production Index Base 100"]
illinois2 = data["Public interactivity for Illinois"]
pearsonr_illinois, p_value = pearsonr(illinois1, illinois2)
print("The Pearson coefficient for Illinois is :", pearsonr_illinois)

indiana1 = data["Indiana GM Production Index Base 100"]
indiana2 = data["Public interactivity for Indiana"]
pearsonr_indiana, p_value = pearsonr(indiana1, indiana2)
print("The Pearson coefficient for Indiana is :", pearsonr_indiana)

iowa1 = data["Iowa GM Production Index Base 100"]
iowa2 = data["Public interactivity for Iowa "]
pearsonr_iowa, p_value = pearsonr(iowa1, iowa2)
print("The Pearson coefficient for Iowa is :", pearsonr_iowa)

michigan1 = data["Michigan GM Production Index Base 100"]
michigan2 = data["Public interactivity for Michigan"]
pearsonr_mich, p_value = pearsonr(michigan1, michigan2)
print("The Pearson coefficient for Michigan is :", pearsonr_mich)

minnesota1 = data["Minnesota GM Production Index Base 100"]
minnesota2 = data["Public interactivity for Minnesota"]
pearsonr_min, p_value = pearsonr(minnesota1, minnesota2)
print("The Pearson coefficient for Minnesota is :", pearsonr_min)

missouri1 = data["Missouri GM Production Index Base 100"]
missouri2 = data["Public interactivity for Missouri"]
pearsonr_mis, p_value = pearsonr(missouri1, missouri2)
print("The Pearson coefficient for Missouri is :", pearsonr_mis)

nebraska1 = data["Nebraska GM Production Index Base 100"]
nebraska2 = data["Public interactivity for Nebraska"]
pearsonr_neb, p_value = pearsonr(nebraska1, nebraska2)
print("The Pearson coefficient for Nebraska is :", pearsonr_neb)

ohio1 = data["Ohio GM Production Index Base 100"]
ohio2 = data["Public interactivity for Ohio"]
pearsonr_ohio, p_value = pearsonr(ohio1, ohio2)
print("The Pearson coefficient for Ohio is : ", pearsonr_ohio)

texas1 = data["Texas GM Production Index Base 100"]
texas2 = data["Public interactivity for Texas"]
pearsonr_tex, p_value = pearsonr(texas1, texas2)
print("The Pearson correlation for Texas is : ", pearsonr_tex)

wisconsin1 = data["Wisconsin GM Production Index Base 100"]
wisconsin2 = data["Public interactivity for Wisconsin"]
pearsonr_wis, p_value = pearsonr(wisconsin1, wisconsin2)
print("The Pearson correlation for Wisconsin is : ", pearsonr_wis)

#We now create a barchart to see all correlations levels between states
#We create the X and Y variables
allpearson = [pearsonr_illinois, pearsonr_indiana, pearsonr_iowa, pearsonr_mich, pearsonr_min, pearsonr_mis, pearsonr_neb, pearsonr_ohio, pearsonr_tex, pearsonr_wis]
states = ["Illinois","Indiana","Iowa","Michigan","Minnesota","Missouri","Nebraska","Ohio","Texas","Wisconsin"]

#We create the bar chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(states, allpearson)

#We show the treshold lines
plt.axhline(y=0.7,linewidth=1, linestyle = "-", label = "Highly correlated")
plt.axhline(y=0.5,linewidth=1, linestyle = "-.", label = "Moderatly correlated")
plt.axhline(y=0.3,linewidth=1, linestyle = "--", label = "Poorly correlated")
l = plt.legend(loc ='upper right')

plt.show()
```
